```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

alpha = 0.4
n, m = 7, 7
x, y = np.mgrid[0:n+1,0:m+1]

x1, y1 = 3.5, 2.5
alpha1 = 0.4
alpha2 = 1.
x2, y2 = 6, 6
z = (1.5*np.exp(-(((x - x1)*alpha)**2 + ((y - y1)*alpha1)**2))*9 + 1.5*np.exp(-(((x - x2)*alpha2)**2 + ((y - y2)*alpha2)**2))*9).astype(int)/10
z = z + np.random.normal(0, 0.06, size=(n+1,m+1))
plt.imshow(z, cmap="gray")
mpimg.imsave('bw.png', z, cmap="gray")
```
