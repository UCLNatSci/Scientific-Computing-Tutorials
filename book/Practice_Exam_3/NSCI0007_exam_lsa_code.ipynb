```python
G_2 = nx.petersen_graph()

G_3 = nx.dodecahedral_graph()

G_4 = nx.complete_graph(7)




subax1 = plt.subplot(121)
nx.draw(G_2, with_labels=True, font_weight='bold')
subax2 = plt.subplot(122)
nx.draw_shell(G_2, nlist=[range(5, 10), range(5)], with_labels=True, font_weight='bold')


graphs = [(G_2, "Petersen Graph"), (G_3, "Dodecahedral Graph"), (G_4, "Complete Graph (Order 7)")]

for i in range(len(graphs)):
    with open("graph0{}.txt".format(i), "w") as f:
        title = graphs[i][1]
        G = graphs[i][0]
        A = nx.linalg.graphmatrix.adjacency_matrix(G).toarray()
        f.write(title + "\n")
        f.write(str(A.shape[0]) + "\n")
        for row in A:
            for i in row:
                f.write(str(i))
            f.write("\n")        
```
